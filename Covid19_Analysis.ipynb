#  Italy Covid19 Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('covid19-ita-province-latest.csv',header=0, index_col='date', parse_dates=True)
data.head(3)
```

```python
data.describe()
```

```python
data.info()
```

```python
data.isnull().sum()
```

```python
data['note_en'].fillna(0,inplace=True)
data['note_it'].fillna(0,inplace=True)
data['province_ISO'].fillna(0,inplace=True)
data.isnull().sum()
```

```python
region=data.groupby('region')
x=region['total_cases'].agg(np.mean)
#y=region['province '].agg(np.mean)
plt.plot(x)
plt.xticks(rotation=90)
plt.title('Covid19 Analysis')
plt.xlabel('region')
plt.ylabel('total_cases')
plt.show()
```
